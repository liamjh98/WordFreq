## Step 1 : Importing Packages

````python
import matplotlib.pyplot as plt
%matplotlib inline
import operator
````

## Step 2 : Importing Text

`f` is used as a variable to hold the text, `open` is used to open a file and get it ready to be read from.

````python
f = open("text.txt", "r")
words = f.read();
print(words)
````

## Step 3 : Creating the Dictionary and Plotting Graph

````python
#Takes the words from file and splits them up
wordlist = words.split()
wordfreq = []

#gets the frequency of each word
for w in wordlist:
    wordfreq.append(wordlist.count(w))

#creates a dictionary of words and frequency eg ("Hello", 2)
wordfreq = dict(list(zip(wordlist,wordfreq)))

#removes entries from dictionary that dont have a frequency greater than set value (freq_of_word)
freq_of_word = 5 #will only plot words which occur x times or more

wordfreq = {key:val for key, val in wordfreq.items() if val >= freq_of_word}

wordfreq = dict( sorted(wordfreq.items(), key=operator.itemgetter(1),reverse=True))
print(wordfreq)


#Plot the graph
plt.figure(figsize=(40,10))
bar1 = plt.bar(list(wordfreq.keys()), list(wordfreq.values()), 0.5, 1)
plt.tick_params(axis='x', rotation=70)

fnt_size = 10 #change the font size of the x and y tics

plt.tick_params(axis='x', labelsize=fnt_size)
plt.tick_params(axis='y', labelsize=fnt_size)

plt.gcf().subplots_adjust(bottom=0.15)
plt.gcf().subplots_adjust(left=0.05)

# Add counts above the two bar graphs
for rect in bar1:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%d' % int(height), ha='center', va='bottom')

plt.margins(0.005,0.005) #set the padding for the graph
plt.tight_layout() 
plt.axis('tight')
plt.margins(x=0)

#Change the title, xlabel and ylabel
plt.title("Frequency of Words")
plt.xlabel("Word")
plt.ylabel("Frequency")

plt.savefig('figure.pdf',bbox_inches='tight') #saves the figure as a pdf file
plt.show()
````

## Explaining the above code

Here I will go through each important part of the code and explain how it works

```OCTAVE
for w in wordlist:
    wordfreq.append(wordlist.count(w))
```

1. The **for loop** will use a dummy variable `w` and this will be set to each word in our `wordlist`, `wordfreq` is our array, where the frequency of each word will be stored. Two more functions are used to help us do this; `append` is a function which does just that, adds a new item into the array. Then `wordlist.count(w)` will actually get the frequency of the current word in our text file.


2. Then `wordfreq = dict(list(zip(wordlist,wordfreq)))` will merge the two variables `wordlist` and `wordfreq` into a python **dictionary** which consists of **keys** and **values**. More information about dictionaries can be found here: https://www.w3schools.com/python/python_dictionaries.asp.

```OCTAVE
freq_of_word = 5

wordfreq = {key:val for key, val in wordfreq.items() if val >= freq_of_word}

wordfreq = dict( sorted(wordfreq.items(), key=operator.itemgetter(1),reverse=True))
print(wordfreq)
```

3. The next part of the code filters the words by occurance, by setting the value of `freq_of_word` will only show words which occur that amount of times or more. Then the next line of code, `wordfreq = dict( sorted(wordfreq.items(), key=operator.itemgetter(1),reverse=True))`, will sort the dictionary by the number of words in decending order.


4. The next section is envoled in plotting the graph. Matplotlib is used to do this, and there are many different plots which can be used, here a bar chart is used. The size of the figure can be set using the first line of code. The line of code that actually plots the bar chart is `bar1 = plt.bar(list(wordfreq.keys()), list(wordfreq.values()), 0.5, 1)`, where the keys and the values talked about earlier are set for the x and y values in the first two arguments. `wordfreq` is the variable and `.keys()` is the accessor used to get the keys from the dictionary. 


5. After this the rest of the code changes how the graph looks. `plt.tick_params(axis='x', rotation=70)` changes the x axis labels to be rotated. The font size of all the labels can be adjusted. The graph can be saved to file, `plt.savefig('figure.pdf',bbox_inches='tight')` saves the graph as a pdf file, which can be found in the folder this project was in. The final line of code will output the plot `plt.show()`
